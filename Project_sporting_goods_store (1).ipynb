```python
#Search Product
def search(table):
  try:
      product_id = int(input("Enter product ID: "))
      if product_id in table:
          print(tabulate([table[product_id]], headers=table[0], tablefmt='fancy_grid'))
      else:
          print("ID does not exist.\n")
  except ValueError:
      print("Invalid input. Please enter a valid number.\n")
```

**Task 5: Saving Tables**

Saving Product and customer

```python
def save_table(table, filename):
  table_list = list(table.values())
  with open(f"{filename}.csv", 'w', newline='') as file:
      writer = csv.writer(file)
      writer.writerows(table_list)
  print(f"Table saved to {filename}.csv")

def load_table(filename):
  table = {}
  try:
      with open(f"{filename}.csv", newline='') as file:
          reader = csv.reader(file)
          for idx, row in enumerate(reader):
              if idx == 0:
                  table[0] = row
              else:
                  table[int(row[0])] = row
      print(f"Table loaded from {filename}.csv")
  except FileNotFoundError:
      print(f"No existing file found for {filename}.csv. Starting with an empty table.")
      if filename == 'Product':
        table[0] = ['id', 'name', 'category','price']
      elif filename == 'Customer':
        table[0] = ['id', 'name', 'address','phone']
      elif filename == 'Supplier':
        table[0] = ['id', 'name', 'address','phone']
      elif filename == 'Employee':
        table[0] = ['id', 'name', 'address','phone']
      elif filename == 'Sales Transaction':
        table[0] = ['customer_id', 'employee_id', 'product_id', 'quantity', 'date_of_sales']
  return table
```

Function to select

```python
def productsTable():
  global product_table
  product_table = load_table('Product')
  while True:
      print("1. Add Product\n2. Search Product\n3. Modify Product\n4. Display All Product\n5. Save and Exit")
      choice = input("Enter your choice: ")
      if choice == "1":
          add_product()
      elif choice == "2":
          search(product_table)
      elif choice == "3":
          modify_product()
      elif choice == "4":
          display(product_table)
      elif choice == "5":
          save_table(product_table,'Product')
          print("Data saved. Exiting...")
          break
```
